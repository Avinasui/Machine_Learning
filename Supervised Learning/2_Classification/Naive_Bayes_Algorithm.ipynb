# Train/ Test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train.shape,x_test.shape
```

# Null value treatment

```python
x_train[cat].isnull().mean() # represent the percentage of null value
```

```python
for df2 in [x_train,x_test]:
    df2['work_class'].fillna(x_train['work_class'].mode()[0],inplace=True)
    df2['occupation'].fillna(x_train['occupation'].mode()[0],inplace=True)
    df2['native_country'].fillna(x_train['native_country'].mode()[0],inplace=True)
```

# Encoding - category_encoder

```python
import category_encoders as ce
encoder=ce.OneHotEncoder()

x_train=encoder.fit_transform(x_train)
x_test=encoder.transform(x_test) # always use .trainsform for test column else shape mismath would happen
```

# Feature Scaling

```python
colss=x_train.columns
```

```python
from sklearn.preprocessing import RobustScaler # removes outliers
scaler=RobustScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# Model Definition

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
preds=model.predict(x_test)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report
acc=accuracy_score(y_test,preds)
acc
```

```python
pred_tr=model.predict(x_train)
acc_tr=accuracy_score(y_train,pred_tr)
acc_tr
```
