# Data Preprocessing

```python
#Check for imbalanced output
df.outcome.value_counts()
```

```python
#Null value check
df.isnull().sum()
```

# X and Y declaration

```python
#X and Y declaration
x=df.iloc[:,0:-1]
x
```

```python
y=df['outcome']
y
```

# Train Test Split

```python
#Here we are not scaling becoz tree algorithm works well without scaling
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=41)
```

```python
x_train.shape
y_train.shape
```

# Model-Decision tree

```python
#Model
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier(max_depth=4)#play with parameters use ,
```

```python
model1.fit(x_train,y_train)
```

```python
pred1=model1.predict(x_test)
pred1

pred1_tr=model1.predict(x_train)

compare=pd.DataFrame({"Actual":y_test,"Predicted":pred1})
compare
```

# Evaluation

```python
#Evaluation
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
test_acc=accuracy_score(y_test,pred1)
train_acc=accuracy_score(y_train,pred1_tr)
print("Training Accuracy",train_acc)
print("Testing Accuracy",test_acc)#Decision tree always show overfitting 
#Best accuracy at random state=41
```

```python
pd.crosstab(y_test,pred1)
```

```python
print(classification_report(y_test,pred1)) #High recall then good classification
```
